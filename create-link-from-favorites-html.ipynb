`````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os 
`````

`````python
# Renomei o arquivo se precisar
file = './favoritos_23_10_2021.html'
with open(file) as fp:
    soup = BeautifulSoup(fp, 'html.parser')
`````

`````python
# Busca a se√ß√¢o da pasta 'W'
all_web_section = soup.find_all('h3')

for w in all_web_section:
    if w.get_text() == 'W':
        web_section = w

print(web_section.get_text())
`````

Cada DL √© uma pasta que tem v√°rios A

````
</DL><p>
<DT><H3 ADD_DATE="1633959192" LAST_MODIFIED="1634181324">Random Web Uncategorized</H3>

````

`````python
# Pego todas as pasta, N√ÉO PODE HAVER SUB PASTS, SE N√ÇO ACHO QUE N√ÉO FUNCIONA
list_dl = web_section.parent.find('dt').find_all('dl')
`````

`````python
def get_tuple_from_a_tag(tag):
    # (Link, Text of Link in Favorites)
    return (tag['href'], tag.text)
`````

`````python
all_favorites = {}

for dl in list_dl:
    list_tuples = []
    folder_name = dl.previous_sibling.previous_sibling.get_text()
    print('\n', folder_name, '\n')
    all_links = dl.find_all('a')
    for l in all_links:
        list_tuples.append(get_tuple_from_a_tag(l))
    print(list_tuples)
    all_favorites[folder_name] = list_tuples    
    
`````

`````python
all_favorites
`````

`````python
with open("my-link.js",'w', encoding = 'utf-8') as f:
    for folder_title, list_tuples in all_favorites.items():
        f.write('\n')
        f.write('var ' + folder_title + ' =  `\n')
        for atuple in list_tuples:
            f.write('\t\t\t')
            f.write(atuple[1]) # title
            f.write('\n')
            f.write('\t\t\t')
            f.write(atuple[0]) # link
            f.write('\n')
        f.write('\t\t`\n') # End
        f.write('\n\n')
`````
